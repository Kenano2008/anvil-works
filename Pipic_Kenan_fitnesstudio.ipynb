```python
import sqlite3

# Verbindung zur DB (Datei wird erstellt)
conn = sqlite3.connect("Pipic_Kenan_fitnessstudio.db")
cursor = conn.cursor()

# Tabellen erstellen
cursor.execute("""
CREATE TABLE Trainer (
    TID INTEGER PRIMARY KEY,
    Vorname VARCHAR(50),
    Nachname VARCHAR(50),
    Spezialgebiet TEXT
)
""")

cursor.execute("""
CREATE TABLE Mitglieder (
    MID INTEGER PRIMARY KEY,
    Vorname VARCHAR(50),
    Nachname VARCHAR(50),
    Email TEXT,
    Beitrittsdatum DATETIME
)
""")

cursor.execute("""
CREATE TABLE Kurs (
    KID INTEGER PRIMARY KEY,
    TID INTEGER,
    Bezeichnung TEXT,
    Uhrzeit DATETIME,
    Wochentag TEXT,
    MaxTeilnehmeranzahl INT,
    FOREIGN KEY (TID) REFERENCES Trainer(TID)
)
""")

cursor.execute("""
CREATE TABLE Anmelden (
    MID INTEGER,
    KID INTEGER,
    Anmeldedatum DATETIME,
    PRIMARY KEY (MID, KID),
    FOREIGN KEY (MID) REFERENCES Mitglieder(MID),
    FOREIGN KEY (KID) REFERENCES Kurs(KID)
)
""")

# =========================
# Daten einfügen
# =========================

# Trainer
trainer = [
    (1, "Max", "Müller", "Fitness"),
    (2, "Anna", "Schmidt", "Yoga"),
    (3, "Lukas", "Meier", "Cardio")
]
cursor.executemany("INSERT INTO Trainer VALUES (?, ?, ?, ?)", trainer)

# Mitglieder
mitglieder = [
    (1, "Amar", "Pipic", "amar@mail.com", "2026-01-01"),
    (2, "Lisa", "Huber", "lisa@mail.com", "2026-02-10"),
    (3, "Tom", "Keller", "tom@mail.com", "2026-03-15")
]
cursor.executemany("INSERT INTO Mitglieder VALUES (?, ?, ?, ?, ?)", mitglieder)

# Kurse
kurse = [
    (1, 1, "Krafttraining", "2026-04-10 18:00", "Montag", 10),
    (2, 2, "Yoga Anfänger", "2026-04-11 17:00", "Dienstag", 15),
    (3, 3, "HIIT", "2026-04-12 19:00", "Mittwoch", 12)
]
cursor.executemany("INSERT INTO Kurs VALUES (?, ?, ?, ?, ?, ?)", kurse)

# Anmeldungen
anmeldungen = [
    (1, 1, "2026-04-01"),
    (1, 2, "2026-04-02"),
    (2, 2, "2026-04-03"),
    (3, 3, "2026-04-04")
]
cursor.executemany("INSERT INTO Anmelden VALUES (?, ?, ?)", anmeldungen)

# Speichern
conn.commit()

# =========================
# Testabfrage
# =========================
for row in cursor.execute("""
SELECT m.Vorname, m.Nachname, k.Bezeichnung
FROM Anmelden a
JOIN Mitglieder m ON a.MID = m.MID
JOIN Kurs k ON a.KID = k.KID
"""):
    print(row)

conn.close()
```
